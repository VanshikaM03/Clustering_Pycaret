```python
from pycaret.utils import version
version()
from pycaret.clustering import *
from pycaret.datasets import get_data
import pandas as pd
```

```python
file_path = "/content/ShoppingComplex.csv"
datafile = pd.read_csv(file_path)
```

```python
datafile.head()
```

```python
setup(data = datafile, verbose=False)
print("For Kmeans")
x = create_model('kmeans')
print("For Cluster = 3")
x3 = create_model('kmeans', num_clusters = 3)

print("For Cluster = 4")
x4 = create_model('kmeans', num_clusters = 4)

print("For Cluster = 5")
x5 = create_model('kmeans', num_clusters = 5)

print("For Agglomerative Clustering")
y = create_model('hclust')

print("For Cluster = 3")
y3 = create_model('hclust', num_clusters = 3)

print("For Cluster = 4")
y4 = create_model('hclust', num_clusters = 4)

print("For Cluster = 5")
y5 = create_model('hclust', num_clusters = 5)

print("For Birch")
z = create_model('birch')

print("For Cluster = 3")
z3 = create_model('birch', num_clusters = 3)

print("For Cluster = 4")
z4 = create_model('birch', num_clusters = 4)

print("For Cluster = 5")
z5 = create_model('birch', num_clusters = 5)
```

```python
kmeanpredict = predict_model(x, data = datafile)
kmeanpredict
```

```python
kmeanpredict.to_csv("kmeanprediction.csv")
print("Result file save sucessfully!!")
```
